```python
import numpy as np


for col_name in dataSet1.columns: 
    currentCol = col_name
    myColumns = ["STA", "Precip", "WindGustSpd", "MaxTemp", "MinTemp", "MeanTemp", "Snowfall"]
    listLength = len(myColumns)
    x = 0
    while x <= listLength -1:
        if str(currentCol) == str(myColumns[x]):
            dataSet1[myColumns[x]] = pd.to_numeric(dataSet1[myColumns[x]])
            x += 1
            break
        else:
            x += 1        
```

```python
merged_left = pd.merge(left=dataSet1, right=dataSet2, how='left', left_on='STA', right_on='WBAN')
merged_left.head()
```

```python
def max_lens(df):
    lens = []
    for x in df.columns:
        sublen = []
        for i, row, in df.iterrows():
            sublen.append(len(str(row[x])))
        lens.append(max(sublen))
    return lens
```

```python
max_lens(merged_left)
```

```python
groupedSet = merged_left.groupby('NAME').mean()
groupedSet
```

```python
groupedSet.drop('MeanTemp', inplace=True, axis=1)
groupedSet
```

```python
def meanTempFunc(df):
    meanT = (df['MaxTemp'] + df['MinTemp'])/2
    return round(meanT, 2)
```

```python
groupedSet.apply(meanTempFunc, axis=1)
```

```python
groupedSet['meanTempFunc'] = groupedSet.apply(meanTempFunc, axis=1)
groupedSet
```
